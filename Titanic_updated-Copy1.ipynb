```python
train=train.drop(['Name','Cabin','Ticket','PassengerId'],axis=True)
train
```

```python
train.hist(figsize=[20,20])
plt.show()
```

```python
train.median()
```

```python
train["Age"].fillna(train["Age"].mean(),inplace=True)
train['Embarked'].fillna(train['Embarked'].mode()[0],inplace=True)
```

```python
train['Family']=train['SibSp']+train['Parch']
train=train.drop(['SibSp','Parch'],axis=True)
train
```

```python
def bar_chart(column):
    survived=train[train["Survived"]==1][column].value_counts()
    dead=train[train["Survived"]==0][column].value_counts()
    df=pd.DataFrame([survived,dead])
    df.index=["Survived","Dead"]
    df.plot(kind="bar",figsize=(10,5))
    
bar_chart("Pclass")
bar_chart("Sex")
bar_chart("Family")
bar_chart("Embarked")
```

```python
def box_plot(column):
    train.boxplot(by="Survived",column=[column],grid=True)
```

```python
box_plot("Fare")
box_plot('Family')
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(train.corr(), annot=True, linewidths=0.05)
plt.show()
```

```python
X=train.drop(["Survived"],axis=True)
X
```

```python
y=train["Survived"]
y.shape
```

```python
X['Pclass']=X['Pclass'].astype(str)
```

```python
X_dummies=pd.get_dummies(X)
X_dummies
```

```python
X_dummies=X_dummies.drop(['Sex_female'],axis=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = X_dummies.copy()
X_scaled[['Age','Fare','Family']] = scaler.fit_transform(X[['Age','Fare','Family']])
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=.25,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc= RandomForestClassifier(max_depth=200, max_features=7, min_samples_leaf=5,min_samples_split=7, n_estimators= 2000)
rfc.fit(X_train, y_train)
```

```python
rfc.score(X_train, y_train)
```

```python
pred=rfc.predict(X_test)
```

```python
len(pred)
```

```python
from sklearn.metrics import confusion_matrix

result=confusion_matrix(y_test,pred)
result
```
