```python
dataset['cabin_multiple']=train.Cabin.apply(lambda x: 0 if pd.isna(x) else len(x.split(' ')))
dataset['cabin_let']=train.Cabin.apply(lambda x:str(x)[0])
dataset['num_ticket']=train.Ticket.apply(lambda x: 1 if x.isnumeric() else 0)
dataset['letter_ticket']=train.Ticket.apply(lambda x: ''.join(x.split(' ')[:-1]).replace('.','').replace('/','').lower() if len(x.split(' ')[:-1])>0 else 0)
dataset['name_title']=train.Name.apply(lambda x: x.split(',')[1].split('.')[0].strip())
```

```python
dataset.Age=dataset.Age.fillna(train.Age.mean())
dataset.Fare = dataset.Fare.fillna(train.Fare.median())
```

```python
dataset.dropna(subset=['Embarked'],inplace=True)
```

```python
dataset['norm_fare']=np.log(dataset.Fare+1)
dataset['norm_fare'].hist()
```

```python
dataset.Pclass = dataset.Pclass.astype(str)
```

```python
dummies_dataset = pd.get_dummies(dataset[['Pclass','Sex','Age','SibSp','Parch','norm_fare','Embarked','cabin_let','cabin_multiple','num_ticket','letter_ticket','name_title','train_test']])
```

```python
X_train = dummies_dataset[dataset.train_test == 1].drop(['train_test'], axis = 1)
X_test = dummies_dataset[dataset.train_test == 0].drop(['train_test'], axis = 1)
```

```python
y_train=dataset[dataset.train_test==1].Survived
y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
dummies_dataset_scaled = dummies_dataset.copy()
dummies_dataset_scaled[['Age','SibSp','Parch','norm_fare']]=scale.fit_transform(dummies_dataset_scaled[['Age','SibSp','Parch','norm_fare']])
```

```python
dummies_dataset_scaled
```

```python
X_train_scaled = dummies_dataset_scaled[dummies_dataset_scaled.train_test == 1].drop(['train_test'], axis = 1)
X_test_scaled = dummies_dataset_scaled[dummies_dataset_scaled.train_test == 0].drop(['train_test'], axis = 1)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
lr = LogisticRegression(max_iter = 1000)
cv = cross_val_score(lr,X_train_scaled,y_train,cv = 5)
print(cv)
print(cv.mean())
```

```python
dt = tree.DecisionTreeClassifier(random_state = 1)
cv = cross_val_score(dt,X_train_scaled,y_train,cv = 5)
print(cv)
print(cv.mean())
```

```python
knn = KNeighborsClassifier()
cv = cross_val_score(knn,X_train_scaled,y_train,cv = 5)
print(cv)
print(cv.mean())
```

```python
rf = RandomForestClassifier()
cv = cross_val_score(rf,X_train_scaled,y_train,cv = 5)
print(cv)
print(cv.mean())
```

```python
svc = SVC()
cv = cross_val_score(svc,X_train_scaled,y_train,cv = 5)
print(cv)
print(cv.mean())
```

```python
from sklearn.ensemble import VotingClassifier
vc = VotingClassifier(estimators = [('lr',lr),('dt',dt),('knn',knn),('rf',rf),('svc',svc)])
```

```python
cv = cross_val_score(vc, X_train_scaled, y_train, cv=5)
print(cv)
print(cv.mean())
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
def clf_perf(clf, model_name):
    print(model_name)
    print('Best Score: ',clf.best_score_)
    print('Best Params: ',clf.best_params_ )
```

```python
lr = LogisticRegression()
param_grid = {'max_iter' : [1000],
              'penalty' : ['l1', 'l2'],
              'C' : np.logspace(-4, 4, 20),
              'solver' : ['liblinear']}

clf_lr = GridSearchCV(lr, param_grid = param_grid, cv = 5, verbose = True, n_jobs = -1)
best_clf_lr = clf_lr.fit(X_train_scaled,y_train)
clf_perf(best_clf_lr,'Logistic Regression')
```

```python
knn = KNeighborsClassifier()
param_grid = {'n_neighbors' : [3,5,7,9],
              'weights' : ['uniform', 'distance'],
              'algorithm' : ['auto', 'ball_tree','kd_tree'],
              'p' : [1,2]}
clf_knn = GridSearchCV(knn, param_grid = param_grid, cv = 5, verbose = True, n_jobs = -1)
best_clf_knn = clf_knn.fit(X_train_scaled,y_train)
clf_perf(best_clf_knn,'KNN')
```

```python
svc = SVC(probability = True)
param_grid = tuned_parameters = [{'kernel': ['rbf'], 'gamma': [.1,.5,1,2,5,10],
                                  'C': [.1, 1, 10, 100, 1000]},
                                 {'kernel': ['linear'], 'C': [.1, 1, 10, 100, 1000]},
                                 {'kernel': ['poly'], 'degree' : [2,3,4,5], 'C': [.1, 1, 10, 100, 1000]}]
clf_svc = GridSearchCV(svc, param_grid = param_grid, cv = 5, verbose = True, n_jobs = -1)
best_clf_svc = clf_svc.fit(X_train_scaled,y_train)
clf_perf(best_clf_svc,'SVC')
```

```python
rf = RandomForestClassifier(random_state = 1)
param_grid =  {'n_estimators': [100,500,1000], 
                                  'bootstrap': [True,False],
                                  'max_depth': [3,5,10,20,50,75,100,None],
                                  'max_features': ['auto','sqrt'],
                                  'min_samples_leaf': [1,2,4,10],
                                  'min_samples_split': [2,5,10]}
                                  
clf_rf_rnd = RandomizedSearchCV(rf, param_distributions = param_grid, n_iter = 100, cv = 5, verbose = True, n_jobs = -1)
best_clf_rf_rnd = clf_rf_rnd.fit(X_train_scaled,y_train)
clf_perf(best_clf_rf_rnd,'Random Forest')
```

```python
rf = RandomForestClassifier(random_state = 1)
param_grid =  {'n_estimators': [400,450,500,550],
               'criterion':['gini','entropy'],
                                  'bootstrap': [True],
                                  'max_depth': [15, 20, 25],
                                  'max_features': ['auto','sqrt', 10],
                                  'min_samples_leaf': [2,3],
                                  'min_samples_split': [2,3]}
                                  
clf_rf = GridSearchCV(rf, param_grid = param_grid, cv = 5, verbose = True, n_jobs = -1)
best_clf_rf = clf_rf.fit(X_train_scaled,y_train)
clf_perf(best_clf_rf,'Random Forest')
```

```python
best_rf = best_clf_rf.best_estimator_.fit(X_train_scaled,y_train)
feature_importances = pd.Series(best_rf.feature_importances_, index=X_train_scaled.columns)
feature_importances.nlargest(20).plot(kind='barh')
```

```python
best_lr = best_clf_lr.best_estimator_
best_knn = best_clf_knn.best_estimator_
best_svc = best_clf_svc.best_estimator_
best_rf = best_clf_rf.best_estimator_
```

```python
vc_hard = VotingClassifier(estimators = [('lr', best_lr),('knn',best_knn),('rf',best_rf),('svc',best_svc)], voting = 'hard') 
vc_soft = VotingClassifier(estimators = [('lr', best_lr),('knn',best_knn),('rf',best_rf),('svc',best_svc)], voting = 'soft')
print('vc_hard :',cross_val_score(vc_hard,X_train_scaled,y_train,cv=5))
print('vc_hard mean :',cross_val_score(vc_hard,X_train_scaled,y_train,cv=5).mean())

print('vc_soft:',cross_val_score(vc_soft,X_train_scaled,y_train,cv=5))
print('vc_soft mean :',cross_val_score(vc_soft,X_train_scaled,y_train,cv=5).mean())
```

```python
vc_soft.fit(X_train_scaled, y_train)
vc_hard.fit(X_train_scaled, y_train)
best_rf.fit(X_train_scaled, y_train)
```

```python
y_pred_vc_soft =  vc_soft.predict(X_test_scaled).astype(int)
y_pred_vc_hard = vc_hard.predict(X_test_scaled).astype(int)
y_pred_rf = best_rf.predict(X_test_scaled).astype(int)
```

```python
final_data_soft = {'PassengerId': test.PassengerId, 'Survived':y_pred_vc_soft}
preds_soft = pd.DataFrame(data=final_data_soft)

final_data_hard = {'PassengerId': test.PassengerId, 'Survived': y_pred_vc_hard}
preds_hard = pd.DataFrame(data=final_data_hard)

final_data_rf = {'PassengerId': test.PassengerId, 'Survived': y_pred_rf}
preds_rf = pd.DataFrame(data=final_data_rf)
```

```python
final_data_comp = {'PassengerId': test.PassengerId, 'Survived_vc_hard': y_pred_vc_hard, 'Survived_rf': y_pred_rf, 'Survived_vc_soft' : y_pred_vc_soft}
comparison = pd.DataFrame(data=final_data_comp)
```

```python
comparison
```

```python
comparison['difference_rf_hard'] = comparison.apply(lambda x: 1 if x.Survived_vc_hard != x.Survived_rf else 0, axis =1)
comparison['difference_soft_hard'] = comparison.apply(lambda x: 1 if x.Survived_vc_hard != x.Survived_vc_soft else 0, axis =1)
comparison['difference_rf_soft'] = comparison.apply(lambda x: 1 if x.Survived_vc_soft != x.Survived_rf else 0, axis =1)
```
